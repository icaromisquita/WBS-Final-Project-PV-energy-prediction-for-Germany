```python
import pickle
import pandas as pd
rfr_model = pickle.load(open("Pv_RamdForReg_model.sav", "rb" ))
```

```python
pv_dfbase = pd.read_csv("https://raw.githubusercontent.com/icaromisquita/WBS-Final-Project-/main/pv_final_cleaned.csv")
pv_dfbase.head()
```

```python
temp = pd.DataFrame(pv_dfbase.mean()).transpose()
final = pd.concat([pv_dfbase[['weekday_name','month_name']].mode(),temp],axis=1)
final
```

```python
#Converting month and year to a integer
final = final.astype({"month":"int","year":"int"})
final
```

## Making a prediction using the base values

```python
print('The Predict energy is {0:.2f} Wh'.format(rfr_model.predict(final)[0]))
print('The Predict energy is {0:.2f} KWh'.format((rfr_model.predict(final)[0])/1000))
```

## Saving the base dataframe

```python
final.to_csv("./pv_base.csv",index=False)
```
